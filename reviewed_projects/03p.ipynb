### first_day_exposition, days_exposition, total_images

–ü—Ä–∏–≤–µ–¥–µ–º –¥–∞—Ç—É —Ä–∞–∑–º–µ—â–µ–Ω–∏—è –≤ —Ä–∞–±–æ—á–∏–π –≤–∏–¥

```python
df.first_day_exposition = pd.to_datetime(df.first_day_exposition,
                                         format="%Y-%m-%dT%H:%M:%S")
```

–ò —Å–ø—Ä–æ—Å–∏–º —Å–µ–±—è - –∫–∞–∫—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –¥–∞–µ—Ç –¥–∞—Ç–∞?  
–ö–æ–≥–¥–∞ –æ—Å—É—â–µ—Å—Ç–≤–ª—è–ª–∞—Å—å –ø—Ä–æ–¥–∞–∂–∞.  
–í –∫–∞–∫–æ–π –≥–æ–¥

```python
df["year_exposition"] = df.first_day_exposition.dt.year
```

–í –∫–∞–∫–æ–π –º–µ—Å—è—Ü

```python
df["month_exposition"] = df.first_day_exposition.dt.month_name()
```

–ö–∞–∫ —Ä–∞–∑–≤–∏–≤–∞–ª—Å—è —Å–µ—Ä–≤–∏—Å —è–Ω–¥–µ–∫—Å-–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏?

```python
months = ["January",
          "February",
          "March",
          "April",
          "May",
          "June",
          "July",
          "August",
          "September",
          "October",
          "November",
          "December",]
      
plt.figure(figsize=(15, 9))
plt.title("Days exposition per month")

#—Å–æ–∑–¥–∞–Ω–∏–µ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
df1 = (df.groupby(["year_exposition", "month_exposition"])
    .days_exposition.median()
    .reset_index())

#–¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏ —Å –Ω–æ–º–µ—Ä–∞–º–∏ –º–µ—Å—è—Ü–µ–≤ –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–ø–∏—Å–∫–∞ month –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –º–µ—Å—è—Ü–µ–≤
df1["m_num"] = df1.month_exposition.apply(lambda x: months.index(x))
df1 = df1.sort_values(["year_exposition", "m_num"])
df1.days_exposition.plot(kind="bar", sort_columns=False, rot=80).set(ylabel=("Days"))
#—Å–æ–∑–¥–∞–Ω–∏–µ –ø–æ–¥–ø–∏—Å–µ–π –æ—Å–∏ —Ö
plt.xticks(np.arange(54), df1.year_exposition.astype(str) + df1.month_exposition);
```

–ù–∞ –∑–∞—Ä–æ–∂–¥–µ–Ω–∏–∏, –æ–Ω —è–≤–Ω–æ –Ω–µ –ø–æ–ª—å–∑–æ–≤–∞–ª—Å—è –¥–æ–≤–µ—Ä–∏–µ–º. –ù–æ —Å–Ω–∏–∂–∞—é—â–µ–µ—Å—è –≤—Ä–µ–º—è –∑–∞–∫—Ä—ã—Ç–∏—è –æ–±—ä—è–≤–ª–µ–Ω–∏–π –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–∞–∑–≥–æ–Ω –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏.

```python
plt.figure(figsize=(15, 9))
plt.title("Close per mount")
#—Å–æ–∑–¥–∞–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
#–¥–µ–Ω—å –∑–∞–∫—Ä—ã—Ç–∏—è —Å—á–∏—Ç–∞–µ—Ç—Å—è –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–Ω—è —Å–æ–∑–¥–∞–Ω–∏—è –æ–±—ä—è–≤–ª–µ–Ω–∏–µ+–¥–Ω–∏ –¥–æ —Å–Ω—è—Ç–∏—è
df1 = pd.DataFrame(df[~df.days_exposition.isna()].apply(
        lambda x: x.first_day_exposition + pd.Timedelta(days=x.days_exposition), axis=1),
        columns=["last_day"],)
#–¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –≥–æ–¥–∞, –º–µ—Å—è—Ü–∞ –∏ –Ω–æ–º–µ—Ä–∞ –º–µ—Å—è—Ü–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
df1["year"] = df1.last_day.dt.year
df1["month"] = df1.last_day.dt.month_name()
df1 = df1.groupby(["year", "month"]).agg({"last_day": "count"}).reset_index()
df1["m_num"] = df1.month.apply(lambda x: months.index(x))
df1 = df1.sort_values(["year", "m_num"])
df1.last_day.plot(kind="bar", sort_columns=False, rot=80, grid=True).set(ylabel=("Counts"))
plt.xticks(np.arange(len(df1)), df1.year.astype(str) + df1.month);
```

```python
years = [2015, 2016, 2017, 2018, 2019]
f, axs = plt.subplots(12, len(years), figsize=(20, 40))
plt.subplots_adjust(hspace=0.6)
axs = axs.ravel()
i = -1 #–∏–Ω–¥–µ–∫—Å –¥–ª—è –∑–∞–ø–æ–ª–¥–µ–Ω–∏—è subplots –ø–æ –≥–æ–¥–∞–º
for month in months:
    for year in years:
        df1 = df[(df.month_exposition == month)
            & (df.year_exposition == year)
            & (~df.days_exposition.isna())]
        i += 1
        title = month
        if df1.shape[0] > 10:
            if i <= len(years):
                title = f"{year}\n{month}"
            sns.scatterplot(ax=axs[i],
                            x=df1.days_exposition,
                            y=df1.total_images).set_title(title)
```

–ü–æ—á–∏—Å—Ç–∏–º –¥–∞–Ω–Ω—ã–µ –≤ —ç—Ç–∏ –¥–Ω–∏, –Ω–æ –∞–∫–∫—É—Ä–∞—Ç–Ω–æ. –£–¥–∞–ª–∏–º —Ç–æ–ª—å–∫–æ —Ç–µ, —É –∫–æ—Ç–æ—Ä—ã—Ö —ç—Ç–∏ "–¥–Ω–∏ —Å–º–µ—Ä—Ç–∏" –ø—Ä–µ–≤—ã—à–∞—é—Ç –º–µ–¥–∏–∞–Ω—É

```python
for year in [2015, 2016, 2017, 2018, 2019]:
    for month in months:
        df = df.drop(labels=df[(df.month_exposition == month) & (df.year_exposition == year)]
            .query(
                "(days_exposition.quantile(0.05)>days_exposition) | (days_exposition>days_exposition.quantile(0.95))"
                  ).index, axis=0)
        # –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Å—è—Ü–∞ —É–±–∏—Ä–∞–µ–º 5% —Å–∞–º—ã—Ö –±—ã—Å—Ç—Ä—ã—Ö, —Å—á–∏—Ç–∞—è –∏—Ö –Ω–µ–¥–æ–æ—Ü–µ–Ω–µ–Ω—ã–º–∏
        # –∏ 5% —Å–∞–º—ã—Ö –¥–æ–ª–≥–∏—Ö, —Å—á–∏—Ç–∞—è –ø–µ—Ä–µ–æ—Ü–µ–Ω–µ–Ω–Ω—ã–º–∏
        for days in [90, 60, 45, 30]:
            if (df[(df.month_exposition == month) & (df.year_exposition == year)
                ].days_exposition.median() < days):
                df = df.drop(
                    labels=df[(df.month_exposition == month)
                                & (df.year_exposition == year)
                                & (df.days_exposition == days)].index,
                    axis=0)
            # —É–¥–∞–ª—è–µ–º –∏–∑ –≤—ã–±–æ—Ä–∫–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è –∏–∑ –¥–Ω–µ–π –∫–æ–Ω—Ü–æ–≤ —Å—Ä–æ–∫–æ–≤ –ø—É–±–ª–∏–∫–∞—Ü–∏–π,
            # –µ—Å–ª–∏ –º–µ–¥–∏–∞–Ω–∞ –∂–∏–∑–Ω–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è –¥–ª—è —ç—Ç–æ–≥–æ –º–µ—Å—è—Ü–∞ –º–µ–Ω—å—à–µ —ç—Ç–æ–≥–æ —Å—Ä–æ–∫–∞
            else:
                break
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è —Ç–∏–ø–æ–≤ –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤
def repl(string):
    replace_value = [
        "–≥–æ—Ä–æ–¥—Å–∫–æ–π –ø–æ—Å–µ–ª–æ–∫ ",
        "–≥–æ—Ä–æ–¥—Å–∫–æ–π –ø–æ—Å—ë–ª–æ–∫ ",
        "–¥–µ—Ä–µ–≤–Ω—è ",
        "–ø–æ—Å–µ–ª–æ–∫ —Å—Ç–∞–Ω—Ü–∏–∏ ",
        "–ø–æ—Å—ë–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞ –∏–º–µ–Ω–∏ ",
        "–ø–æ—Å—ë–ª–æ–∫ –ø—Ä–∏ –∂–µ–ª–µ–∑–Ω–æ–¥–æ—Ä–æ–∂–Ω–æ–π —Å—Ç–∞–Ω—Ü–∏–∏ ",
        "–ø–æ—Å—ë–ª–æ–∫ —Å—Ç–∞–Ω—Ü–∏–∏ ",
        "—Å–∞–¥–æ–≤–æ–¥—á–µ—Å–∫–æ–µ –Ω–µ–∫–æ–º–º–µ—Ä—á–µ—Å–∫–æ–µ —Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ ",
        "—Å–∞–¥–æ–≤–æ–µ —Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤–æ ",
        "—Å–µ–ª–æ ",
        "–ø–æ—Å—ë–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞ ",
        "–ø–æ—Å–µ–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞ ",
        "–∫–æ—Ç—Ç–µ–¥–∂–Ω—ã–π –ø–æ—Å–µ–ª–æ–∫ ",
        "–ø–æ—Å–µ–ª–æ–∫ ",
        "–ø–æ—Å—ë–ª–æ–∫ ",
    ]
    for value in replace_value:
        if string.find(value) == 0:
            return string.replace(value, "")
    return string
```

–ü–µ—Ä–µ–¥–µ–º –∫ –Ω–µ—Ñ–æ—Ä–º–∞–ª—å–Ω–æ–º—É –æ–±—â–µ–Ω–∏—é - –±–µ–∑ —Ç–∏—Ç—É–ª–æ–≤

–û–±—â–∞—è –ø–ª–æ—â–∞–¥—å, –∂–∏–ª–∞—è, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–º–Ω–∞—Ç, –≤—Å–µ —ç—Ç–æ –ø–æ —Å—É—Ç–∏ –≥–æ–≤–æ—Ä–∏—Ç –æ–± –æ–¥–Ω–æ–º –∏ —Ç–æ–º –∂–µ, –æ —Ä–∞–∑–º–µ—Ä–µ –∫–≤–∞—Ä—Ç–∏—Ä—ã. –í—Å–µ —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ –∏–º–µ—é—Ç –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –ª–∏–Ω–µ–π–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å. –ü–æ—ç—Ç–æ–º—É –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –∏—Ö –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å—é, —á—Ç–æ–±—ã –¥–ª—è –Ω–∞—á–∞–ª–∞ –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏

–ü–µ—Ä–µ–¥ –Ω–∞—á–∞–ª–æ–º —É–¥–∞–ª–∏–º —Ç–µ —Å–∞–º—ã–µ –∞–Ω–æ–º–∞–ª–∏–∏ —Å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π –ø–ª–æ—â–∞–¥—å—é.

```python
df = df[(~df.index.isin(df[df.kitchen_area + df.living_area > df.total_area].index))]
```

–ù–∞–π–¥–µ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –æ–±—â–µ–π –ø–ª–æ—â–∞–¥—å—é –∏ –∂–∏–ª–æ–π

```python
#–¥–ª—è –ø–æ–≤—ã—à–µ–Ω–∏—è —Ç–æ—á–Ω–æ—Å—Ç–∏ —Ä–∞—Å—á–µ—Ç–æ–≤ —Å–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –ø–æ–¥—á–∏—â–∞—è –≤ –Ω–µ–º –ø–æ–≥—Ä–∞–Ω–∏—á–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
clear_df = (
    df.loc[~(df.living_area.isna()), ["total_area", "living_area"]]
    .astype(int)
    .query(
        "total_area.quantile(0.01)<total_area<total_area.quantile(0.99)\
                & living_area.quantile(0.01)<living_area<living_area.quantile(0.99)"))
```

```python
regressor = LinearRegression()
X = clear_df.total_area.values.reshape(-1, 1)
y = clear_df.living_area
regressor.fit(X, y)
print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏ {regressor.score(X, y)}')
```

–ò –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –≤ –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏

```python
df.loc[(df.living_area.isna()), "living_area"] = regressor.predict(
    df.loc[(df.living_area.isna()), "total_area"].values.reshape(-1, 1)).astype(int)
```

–° –ø–ª–æ—â–∞–¥—å—é –∫—É—Ö–Ω–∏ –≤—Å–µ –Ω–µ–º–Ω–æ–≥–æ —Å–ª–æ–∂–Ω–µ–µ.

```python
#–¥–ª—è –ø–æ–≤—ã—à–µ–Ω–∏—è —Ç–æ—á–Ω–æ—Å—Ç–∏ —Ä–∞—Å—á–µ—Ç–æ–≤ —Å–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –ø–æ–¥—á–∏—â–∞—è –≤ –Ω–µ–º –ø–æ–≥—Ä–∞–Ω–∏—á–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
cdf = (df.loc[
        ~(df.living_area.isna()) & ~(df.kitchen_area.isna()),
        ["total_area", "living_area", "kitchen_area"],]
    .astype(int)
    .query(
        "total_area.quantile(0.01)<total_area<total_area.quantile(0.99)\
                & living_area.quantile(0.01)<living_area<living_area.quantile(0.99)\
                & kitchen_area.quantile(0.01)<kitchen_area<kitchen_area.quantile(0.99)"))
```

```python
plt.figure(figsize=(9, 9))
plt.xlabel("No living area")
sns.scatterplot(x=(cdf.total_area - cdf.living_area),
                y=cdf.kitchen_area,
                size=cdf.groupby(cdf.total_area - cdf.living_area).kitchen_area.count(),
                sizes=(10, 150),
                alpha=0.7,)
```

–ö–≤–∞—Ä—Ç–∏—Ä—ã –≤ –∫–æ—Ç–æ—Ä—ã—Ö —Ä–∞–∑–Ω–æ—Å—Ç—å —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –±–æ–ª—å—à–µ 45 –º¬≤. –ù–æ –∏ –∫—É—Ö–Ω–∏ —É–∂–µ —Å–ª–∏—à–∫–æ–º –≤–µ–ª–∏–∫–∏, –ø–æ—ç—Ç–æ–º—É –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –æ—Å—Ç–∞–Ω–æ–≤–∏–ª–∞—Å—å. –ü—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã –Ω–µ –º–Ω–æ–≥–æ—á–∏—Å–ª–µ–Ω–æ. –ó–∞–ø–æ–ª–Ω–∏–º –æ–±—â–µ–π –º–µ–¥–∏–∞–Ω–æ–π –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞

```python
rdf = cdf.groupby((cdf.query("13 <(total_area-living_area) < 43").total_area
                    - cdf.query("13 < (total_area-living_area) < 45").living_area)
                 ).kitchen_area.mean()
```

```python
reg = LinearRegression()

X = rdf.index.values.reshape(-1, 1)
y = rdf.values

reg.fit(X, y)
print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏ {reg.score(X, y)}')
print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç {reg.coef_}')
print(f'–ò–Ω—Ç–µ—Ä—Ü–µ–ø—Ç {reg.intercept_}')
```

–î–ª—è –∫–≤–∞—Ä—Ç–∏—Ä—ã-—Å—Ç—É–¥–∏–∏ –∑–∞–ø–æ–ª–Ω–∏–º –ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏ –Ω—É–ª–µ–º, –¥–ª—è –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤ —É—Å–ª–æ–≤–∏—è–º–∏ –æ–ø–∏—Å–∞–Ω–Ω—ã–º–∏ —Ä–∞–Ω–µ–µ.

```python
df.kitchen_area = df.kitchen_area.fillna(
    pd.Series(
        index=df.index,
        data=np.select([df.studio,
                        ((df.total_area - df.living_area) < 13),
                        ((df.total_area - df.living_area) > 45)],
                       [0,
                        df[(df.total_area - df.living_area) < 13].kitchen_area.median(),
                        df[(df.total_area - df.living_area) > 45].kitchen_area.median()],
                       default=reg.predict(
                            np.diff(df.loc[:, ["living_area", "total_area"]], axis=1)
                                            ).astype(int))))
```

```python
df.kitchen_area = df.kitchen_area.round().astype(int)
```

```python
df_coord = pd.DataFrame(city_nan, columns=["locality_name"])
```

```python
def coord_req(city):
    url = "https://geocode-maps.yandex.ru/1.x/?format=json&apikey=c0d403ab-e5be-4049-908c-8122a58acf23&geocode="
    response = r.get(url + "–õ–µ–Ω–∏–Ω–≥—Ä–∞–¥—Å–∫–∞—è –æ–±–ª–∞—Å—Ç—å " + city).json()
    answer = response["response"]["GeoObjectCollection"]["featureMember"][0]["GeoObject"]["Point"].get("pos")
    return tuple(float(x) for x in answer.split(" ")[::-1])
```

–° –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ –∏—Ö —Ü–µ–Ω—Ç—Ä–∞

```python
df_coord["coordinate"] = df_coord.locality_name.apply(coord_req)
```

### prediction model

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞—Å–∫–æ–ª—å–∫–æ –≤–µ—Ä–Ω–æ –±—ã–ª–∏ –≤—ã–±—Ä–∞–Ω—ã –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –¥–ª—è —á–µ–≥–æ –æ–±—É—á–∏–º –ø—Ä–æ–±–Ω—É—é –º–æ–¥–µ–ª—å –æ—Ü–µ–Ω–∏–≤–∞—Ç—å —Å—Ç–æ–∏–º–æ—Å—Ç—å –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞.

```python
X = pd.get_dummies(df.drop(["last_price", "price_per_metr"], axis=1))
y = df.price_per_metr
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=20)
```

```python
rfr = RandomForestRegressor(max_depth=13,
                            n_estimators=200,
                            min_samples_split=10,
                            min_samples_leaf=4)

rfr.fit(X_train, y_train)

display(rfr.score(X_train, y_train))
display(rfr.score(X_test, y_test))
```

```python
#c–æ–∑–¥–∞–µ–º –≤—Ä–µ–º–µ–Ω–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
tr = pd.DataFrame(rfr.predict(X_test), y_test).sort_index().reset_index()
tr.columns = ["real", "predict"]
tr.predict = tr.predict.astype(int)
tr["rel_err"] = ((tr.real - tr.predict) / tr.real * 100).abs()
```

```python
tr.groupby(tr.real.round(-1))\
  .rel_err.mean()\
  .plot(kind="bar", figsize=(10, 6))\
  .set(title="relative error", xlabel=("area"), ylabel=("%"))

plt.axhline(y=tr.rel_err.median(), c="dodgerblue")
```

```python
tr.rel_err.describe([0.9, 0.99])
```
